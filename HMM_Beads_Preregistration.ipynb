## 0) Background

The goal of this study is to provide a principled measure of "optimality" in human decision-making that explicitely takes into account the strategy of the decision maker. There is a wealth of research that proposes that humans make decisions that are at odds with rational, optimal observers. Examples such as framing effects, violations of transitivity, and conjunction fallacies are all prominent examples of cases in which people make decisions that deviate from the most accurate policy or strategy.

Our interest is primarily in the domain of inference, situations in which people use noisy observations to form "mental" models of latent processes in the world (e.g., using past trends to predict the weather tomorrow, or to pick investment strategies). People often perform suboptimally in inference environments, such that the decisions they make are less accurate than ideal observers. For example, a prominent finding is that people "probability match", tending to make predictions that match the actually probability that the events they are predicting will occur (e.g., if trying to predict the next flip from a coin that flips "heads" 80% of the time, people will predict "heads" 80% of the time; see [Vulkan, 2002](https://onlinelibrary.wiley.com/doi/abs/10.1111/1467-6419.00106)). In these environments, the most accurate strategy is to identify the most likely outcome (e.g., "heads" if a coin in 80% biased towards heads) and exclusively predict this outcome (e.g., exclusively guess "heads", resulting and average 80% correct "heads" predictions). Although some people do adopt these types of "maximizing" strategies, the majority adopt less accurate "matching" strategies (e.g., reslting in 64% correct predictions in our coin example). 

Thus, optimality is often measured in terms of accuracy, with strategies that are more accurate being considered more optimal. However, this definition omits an important dimension that we know plays a role in human decision-making: the cognitive effort required to implement the optimal strategy. Optimal solutions can often incure high computational costs, even if they provide marginal improvements in accuracy ([Tavoni et al., 2019](https://www.sciencedirect.com/science/article/pii/S235215461930066X); [Findling et al., 2020](https://www.nature.com/articles/s41593-019-0518-9)). As such, a proper definition of optimality should include not only accuracy, but tradeoff between cost and accuracy.

This notion is at the core of the theory of "bounded rationality" ([Simon, 2000](https://link.springer.com/article/10.1007%252FBF02512227); [Gigerenzer & Selten, 2002](https://pure.mpg.de/rest/items/item_2102373/component/file_2102372/content)). This framework proposes that seemingly irrational or suboptimal behavior occurs due to resource limitations (e.g., computational costs), whereby people look for "good enough" type strategies that acheive an acceptable levels of accuracy. While these strategies likely do not maximize the maximum acheivable accuracy, they may provide enough accuracy in exchange for lower resource costs, and therefore be optimal with respect to the agent's goal of minimizing resource use. This idea has provided a strong foundation for shifting thoughts about what consitutes "optimality" in a decision-making context.

However, while the framework of "bounded rationality" offers an important consideration for measures of optimal behavior, it does not provide a formal framework to assess trade-offs between effort/resources use and accuracy. Here we propose that the information bottleneck method provides such a framework ([Tishby et al., 2001](https://arxiv.org/abs/physics/0004057)). The inforamtion bottleneck method is an information theoretic analysis of trade-offs between the amount of information a model encodes (effort) and the model's predictive performance (accuracy). This method has successfully been applied to a number of different disciplines (e.g., [machine learning](https://ieeexplore.ieee.org/abstract/document/7133169?casa_token=ZeWKoYj5x1EAAAAA:kfKTT6gatuUGqZjfXrO-FAFkWFjg4MlTSCJKwKs6ltaNCCs80aQs8Stq7nCe7Es5dnNCRXt0), [biology](https://www.pnas.org/content/112/22/6908.short), etc.) to specifically measure performance trade-offs. As such, we propose using this method in the domain of human learning to assess how optimally humans use information from their environment to make predictions.

## 1) Hypotheses

With this background, the main goal of this project will be to adapt and apply the information bottleneck method to human behavioral tasks, with the specific goal of identifying how optimally humans trade-off effort and accuracy during inference. Our specific implementation of the method is expanded on in detail in section 3. For all hypotheses, we will be assessing the relationship between the amount of information a person is encoding (their information complexity) and the accuracy of their strategy (predictive accuracy). We will then be comparing where they fall in this complexity/accuracy space to the true optimal information bottleneck bound. People that fall closer to/farther from this bound will be measured as making more/less optimal use of the information they are encoding.

We will use this method to test three primary hypotheses:

### 1.1) Probability matching-like behavior results from optimal tradeoff between complexity and accuracy

We hypothesize that "probability matching" strategies require less information compelxity than "maximizing" strategies. As a result, these probability matching strategies will not be as accurate as maximizing strategies, but we still be optimal, insofar as they make maximally effective use of the reduced information being encoded. We hypothesize that while people may show varying degrees of probability matching-like behavior, they will be similarly close to the maximum acheivable accuracy given the amount of past information they choose to encode. There are 3 specific hypotheses we will test:

** future tense **

* **1) People who use "maximizing"-like strategies will have higher information complexity than "matching" strategies.** We predict that people who "maximize" will use strategies that encode more information about the past than people who probability match.

* **2) People using more information complex strategies will have higher accuracy.** We predict that overall, people that encode more information will be more accurate, given that there is a higher ceiling on their maximum acheivable accuracy.

* **3) People will fall equally far from the optimal bound, regardless of their strategy/information complexity.** We predict that simpler "matching" strategies will make equally effective use of the relatively smaller information being encoded as more information complex "matching" strategies.

The task used to test these hypotheses is outlined in section 2.2.1.

### 1.2) Inference is more optimal in predictibly stable environments than in predictably unstable environments

We have previously found an asymmetry in human inference processes, such that humans tend to learn well in stable environments, when contigencies do not change often (low hazard rate environments), but not well in unstable environments, when contigencies change frequently but in manner that is equally predictable compared to more stable environments (high hazard rate environments; [Filipowicz et al., 2020](https://elifesciences.org/articles/57872)). Pilot analyses on the data from Filipowicz et al., 2020 suggest that this reduction in accuracy resuts from less optimal use of the available information in predictably unstable environments. Moreover, these pilot analyses propose that people using more complex beliefs in unstable environments favor strategies that use information from the immediately preceding trial, rather than from trials farther in the past, as would be expected from more optimal models. There are therefore 4 specific hypotheses we will test:

* **1) People will use less complex beliefs in predictably unstable environments than in stable environments.** Based on pilot analyses on the data from Filipowicz et al., 2020, people tend to use less of the available information in predictably unstable environments than in stable environments.

* **2) People will use information optimally in predictably stable environments.** Pilot analyses on data from Filipowicz et al., 2020 show that while accuracy varies across subjects in relatively stable environments, subjects make equally effective use of the information they encode, no matter how complex.

* **3) People will use information suboptimally in predictably unstable environments because they follow "one-back" strategies.** Pilot analyses on data from Filipowicz et al., 2020 propose that people make less effective use of the information they encode in predictably unstable environments because subjects encode inforamtion primarily about the last observed data-point, which is a suboptimal strategy. We therefore expect that subjects will become increasingly suboptimal as they use more complex strategies.

* **4) People will use information more optimally if they are cued to the correct strategy.** We predict that some of the suboptimalities in the unstable environments stems from strategies that are not appropriate for the particular environment. We therefore expected that more explicit cues about the correct strategies in these environments would improve how optimally people use the information they encode.

### 2.2) Study design

To test our stated hypotheses, we will have participants perform a bead estimation task. In the task, participants will see a white or black bead drawn from one of two jars. Participants will be told the proportion of white/black beads in each jar, and informed how often the jars the beads will be pulled from will switch. A participant's goal will be to use the history of beads they saw on previous trials, and information about jar switch frequency, to predict the color of the next bead as accurately as possible. Using this environment we will have three different experiments.

#### 2.2.1) Markov bead prediction task

This first experiment is meant to test the hypothesis outlined in section 1.1. The first experiment will have 70 participants predict the next bead in an environment where one jar contains 100% black and the other contains 100% white beads. Participants will perform two blocks of 500 trials, with different jar switch rates, or "hazard" rates, in each block. One block will be a "low hazard" condition, in which the jar switch probability will be .32. The other block will be a "high hazard" condition, in which the jar switch probability will be .68. At the start of each task block, participants will be explicitely told the hazard rate, and see two jars with 100% black/white beads throughout the experiment. The order of the high and low hazard rate conditions will be counterbalanced between subjects. Additionally, although the sequence of beads will be generated randomly for each subject, the exact position of the low probability events (jar switches in the low hazard condition and jar repeats in the high hazard condition) will be kept identical, thus preserving the predictability between the two environments and reducing performance differences due to random variations in the different sequences.

These hazard rates were calculated the match the overall mean number of trial-to-trial bead switches expected in the other task environments, which will have hazard rates of .01/.99 in the low and high hazard conditions respectively, and jar probabilities of 80% (identical to the high and low hazard conditions in Filipowicz et al., 2020). Given a hazard rate $H$ and a jar probability $\theta$, the probability that the color $c$ on the current trial $t$ will be the same as the previous trial $t-1$ is computed as:

$$p(c_{t-1}=c_t)=\theta[(1-H)\theta+H(1-\theta)]+(1-\theta)[h\theta+(1-H)(1-\theta)]$$

The average switch probability is computed as:

$$p(c_{t-1}\neq c_t)=1-p(c_{t-1}=c_t)$$

The optimal strategy in this task is to focus on the previous bead, and use your knowledge of the hazard rate to either predict the same color bead (in the low hazard condition), or predict the opposite color bead (in the high hazard condition). Any strategy that does not either always repeat or always switch does not maximize the possible amount of reward and would be indicative of probability matching-like strategies.

In this environment, we will examine whether people do indeed probability match (i.e., switch their predictions in proportion to the number of switches in the condition), whether performance is identical in both hazard rate conditions, whether people's information complexity correlates with predictive accuracy (see section 3.2), and how far people fall from the optimal bound as a function of their information complexity (see section 3.3).

### 3.1) Empirical estimate of the information bottleneck

The problem of building effective mental models can be cast as a problem of "lossy compression": given certain encoding limitations, agents compress past information from their environment into "mental models" that they use to interact with the world. A substantial challenge when trying to measure the "optimality" of these mental models is that the functional form of this compression requires knowing the mental model the agent has formed, something we do not have explicit access to. Furthermore, knowing how effectively this model is using its compressed information is also challenging as it requires knowing the optimal oracle for every given level of compression.

The information bottleneck framework offers solutions to both of these problems. This section will go over the information bottleneck method and how we will apply it to our planned experiments.

The goal of the information bottleneck method is to find the best tradeoffs between compression $\hat{X}$ of one random variable $X$ to predict another $Y$ given their joint distribution $p(X,Y)$. This compression can be obtained by solving the following equation:

$$L[p(\hat{X}|X)]=I(X;\hat{X})-\beta I(\hat{X};Y)$$

where $I(;)$ indicates the mutual information, $L[p(\hat{X}|X)]$ indicates the quatitity to be minimized, and $\beta$ is a Lagrange multiplier. Here, $I(X;\hat{X})$ corresponds to the amount of information our compression $\hat{X}$ retains about $X$, and $\beta I(\hat{X};Y)$ corresponds to the information we want to preserve about $Y$. The Lagrange multiplier $\beta$ controls the amount of information we want to preserve about $Y$. Solving this equation for a range of values of $\beta$, we obtain an information bottleneck curve that tells us the minimum compression we will need to obtain different levels of information about $Y$. When $\beta \rightarrow 0$, none of the past needs to be retained (since we don't need to retain any information about $Y$). Conversely, $\beta \rightarrow \infty$ implies that all of the past is being retained (i.e., no compression), and the curve saturates at $I(X;Y)$.

Typically, the equations above are hard to solve analytically. However, a numerical solution, used in rate distortion theory, can be obtained using the Blahut-Arimoto algorithm. For a source $X$, this algorithm converges on an optimal compression $\hat{X}$ that minimizes the source's distortion $d(X,\hat{X})$ using the joint distribution $p(X,\hat{X})$. It is implemented by iterating between these two equations until convergence:

$$p_{t+1}(\hat{X})=\sum_{x \in X}{p_t(x)p_t(\hat{x}|x)}$$

$$p_{t+1}(\hat{X}|X)=\frac{p_t(\hat{x})\exp(-\beta d_t(x,\hat{X}))}{\sum_{\hat{x} \in \hat{X}}{p_t(\hat{x})\exp(-\beta d_t(x,\hat{x}))}}$$

where $\beta$ indicates the targeted slope in the rate distortion curve (i.e., the amount of distortion we are willing to accept in our signal).

In the context of the information bottleneck, $\beta$ has a similar function, indcating the amount of the variable $X$ we would need to retain a certain amount of inforamtion about $Y$. Tishby and colleagues demonstrate that the distortion $d(x,\hat{x})$ is equivalent to $D_{KL}[p(Y|X)\| p(Y|\hat{X})]$, where $D_{KL}$ is the KL divergence. Therefore, in order to obtain the minimum $p(\hat{X}|X)$ that maximizes $\beta I(\hat{X};Y)$ for a given value $\beta$, we can iterate through the following equations

$$p(\hat{X}|X) = \frac{p(\hat{X})}{Z(X,\beta)}\exp\Big[-\beta\sum_{y \in Y}{p(y|X)\log\frac{p(y|X)}{p(y|\hat{X})}}\Big]$$

where $Z(X,\beta)$ represents the normalizing factor $Z(X,\beta) = \sum_{\hat{x} \in \hat{X}}{p(\hat{x}) \exp \Big[\sum_{y \in Y}{p(y|x) \log \frac{p(y|x)}{p(y|\hat{x})}}\Big]}$,

$$p(Y|\hat{X}) = \frac{1}{p(\hat{x})} \sum_{x \in X}{p(y|x)p(\hat{x}|x)p(x)}$$

to ensure the markovian relation $\hat{X} \leftarrow X \leftarrow Y$ between past, compression and future, and 

$$p(\hat{X}) = \sum_{x \in X}{p(\hat{x}|x)p(x)}$$

Iterating through these three equations provides a distribution $p(\hat{X}|X)$ for a given value of $\beta$ that minimizes $I(X;\hat{X})$, while maximizing $\beta I(\hat{X};Y)$.

To estimate the inforamation bottleneck for our data, we will use a python package we have written ([EMBO](https://pypi.org/project/embo/)) that empirically estimates the information bottleneck curve between two discrete sets of observations.

### 3.2) Applying the information bottleneck to human decision-making data

The information bottleneck method is trying to solve a very similar problem as humans in decision-making tasks: extract the most relevant information from the past, which I will denote $X_{past}$, to form some mental model $M$ that makes accurate predictions about the future $X_{future}$. The amount of information people are encoding from the past $I_{past}$ can be measured as the mutual information between past stimulus features and a participant's mental model:

$$I_{past} \equiv I(X_{past};M) \equiv \sum_{x}\sum_{m} p(x,m) \log_2\Big[\frac{p(x,m)}{p(x)p(m)}\Big]$$

Likewise, the accuracy of the mental model can be measured as the information the mental model carries about the task feature it is trying to predict:

$$I_{future} \equiv I(M; X_{future}) \equiv \sum_{x}\sum_{m} p(m,x) \log_2\Big[\frac{p(m,x)}{p(m)p(x)}\Big]$$

In order to measure these quantities we will need to define the parts of the task that correspond to subject "mental models", "past" events ($X_{past}$), and "future" events ($X_{future}$).

#### 3.2.1) Proxy for "mental models"

We will use subject trial-by-trial responses $R$ as proxies for their mental models. The main assumptions with this method is that a) responses are veridical representations of subject mental models, and b) their mental models are stable throughout the period being measured.

#### 3.2.2) Defining "past" features

To assess the amount of information people are encoding from the task, we will measure $I_{past}$ using increasingly large windows of observations into the past. Technically this window could extend into the entire past, which would make our estimates of the mutual information unfeasible for both measuring human information complexity and empirically estimating the information bottleneck. However, there are a few properties of human memory limits and of the specific sequences we have contructed that make this more tractable.

First, research ont asks such as the "n-back" task show that, due to working memory limits, people are likely not encoding much more than ~3 trials into the past. This makes it more likely that we will be able to capture a substantial portion of the strategies people are likely to use with reasonably limited window sizes into the past.

Second, a property of the sequences we have constructed for our experiment is that the main information they contain can be extracted using a reasonably limited window into the past. This is especially true for the markov sequence with no latent variable, as all of the information needed to predict the future is contained in the previous trial. For this task environment, the information bottleneck only needs to be computed using the previous trial. Similarly, in the latent hidden markov task conditions, the hazard rate serves as a temporal filter, which limits the extent to which past observations are related to future observations. As such, increasing the window size into the past provides deminishing returns related to information about the future. As such, while an infinite window into the past is technically a most correct way to estimate the information bottleneck for a hidden markow environment, it will be well approximated by a more limited window size into the past.

To find the appropriate window to compute $I_{past}$ for both the empirial information bottleneck and subject information complexity, we will compute the mutual information between increasing windows of observations into the past and either future observations ($X_0$) for the information bottleneck, or subject predictions ($R_0$) for subject information complexity.

For the information bottleneck

$$I(X_{-n:-1}; X_0) \equiv \sum_{x_{-n:-1}}\sum_{x_0} p(x_{-n:-1},x_0) \log_2\Big[\frac{p(x_{-n:-1},x_0)}{p(x_{-n:-1})p(x_0)}\Big]$$

and for subject information complexity

$$I(X_{-n:-1}; R_0) \equiv \sum_{x_{-n:-1}}\sum_{r} p(x_{-n:-1},r_0) \log_2\Big[\frac{p(x_{-n:-1},r_0)}{p(x_{-n:-1})p(x_0)}\Big]$$

where $X_{-n:-1}$ corresponds to a window of $n$ previous trials, $X_0$ corresponds to the bead color on the current trial the subject is trying to predict, and $R_0$ correpsonds to the subjects current prediction, before observing $X_0$.

Because we have no specific priors as to where these values should fall, we will use the default uniform priors for all regression coefficients provided by the brms package.

For the information bottleneck in the markov condition, we will just compute the bottleneck considering one trial into the past. In both the hidden markov conditions, and to estimate subject information complexity, we will extend our window into the past as far as our estimates will allow before sampling issues make estimation unfeasible. This will be assed by examining how the mutual information estimates change as a function of window size. To correct for undersampling, the mutual information will be estimated using the [NSB method](https://papers.nips.cc/paper/1965-entropy-and-inference-revisited.pdf) (Nemenman, Shafee, & Bialek). Using this method, when undersampling becomes too problematic, the estimates drop dramatically (see example below). We will therefore compute mutual information values for up to window sizes up until we start seeing these undersampling issues, and use the window size with the highest estimate of the mutual information for our estimates of the empirical information bottleneck bound and subject information complexity.

#### 3.2.3) Defining "future" features

We define future features as the aspect fo the task the subject is trying to predict. In the three tasks above, this corresponds to the color of the next bead. As such, a subject's $I_{future}$ will be computed as

$$I_{future} \equiv I(R_0; X_0) \equiv \sum_{x_0}\sum_{r_0} p(r_0,x_0) \log_2\Big[\frac{p(r_0,x_0)}{p(r_0)p(x_0)}\Big]$$

where $x_0$ indicates the color of the next bead, and $r_0$ indicates a person's prediction about the next bead.

#### 3.2.4) Approximations of mutual information

Mutual information can be heavily affected by the "curse of dimensionality", showing strong positive biases when data are undersampled. To correct for these biases, mutual information will be computed using the [NSB method](https://papers.nips.cc/paper/1965-entropy-and-inference-revisited.pdf) for entropy estimation.

Mutual information estimation error will be computed using bootstrap methods.

### 3.3) Computing subject optimality

In the information bottleneck space, an optimal model is one in which its $I_{future}$ is on the information bottleneck bound for the model's given level of $I_{past}$. To measure how optimally each subject is trading off $I_{past}$ and $I_{future}$, we will measure the vertical distance between each subject's point in the information bottleneck space and the information bottleneck bound. We will use the "distance from the bound" metric to examine the extent to information complexity relates to optimality.

### 3.4) Correlation measures

We plan at least two types of correlations to test our hypotheses: 

1. Correlations between $I_{past}$ and $I_{future}$ to assess the link between complexity and accuracy.

2. Correlations between $I_{past}$ and distance from the bound to measure links between complexity and optimality.

One issue is that both types of correlations depend on $I_{future}$, which has a range that increases as a function of $I_{past}$. This introduces the strong possibility of heteroskedasticity, where variance increases as a function of $I_{past}$. Indeed, this increase in variance is clearly present in our pilot analyses of the data from Filipowicz et al., 2020. 

Additionally, some of our hypothesis predict no correlation between $I_{past}$ and distance from the bound, meaning that we need a framework to measure evidence for the null.

To address both of these issues we will estimate correlations using Bayesian regression implemented in the [Stan](https://mc-stan.org/users/interfaces/rstan) language using the [brms](https://cran.r-project.org/web/packages/brms/index.html) package in R.

We will estimate the posterior estimates for each $\beta$ below:

$$H_1: y=\beta_0+\beta_1I_{past}+N(0,\beta_{\sigma}I_{past})$$

where $y$ will be either $I_{future}$ or distance from the bound, $\beta_0$ corresponds to the intercept (which will measure the offset from the bound in our second correlation), $\beta_1$ measures the influence of $I_{past}$ on our $y$, $N$ indicates gaussian noise, and $\beta_{\sigma}$ captures the heteroskedasticity.

This model will be compared to a simplified intercept only model, which we will use as our null:

$$H_0: y=\beta_0+N(0,\beta_{\sigma}I_{past})$$

To assess whether or not a correlation is present, we will compare the Bayes factor between these models:

$$BF_{10}=\frac{p(D|H_1)}{p(D|H_0)}$$

where $BF_{10}$ indicates the evidence for $H_1$ over $H_0$, and $D$ indicates the observed data. With this formulation, $BF_{10}$ > 1 will indicate evidence for the correlation model $H_1$ and $BF_{10}$ < 1 indicating evidence for the null model $H_0$. As suggested by [Lee et al., 2013](https://psycnet.apa.org/record/2014-14454-000), we will treat $BF_{10}$ values between 3-10 or below 1/3-1/10 as moderate evidence, and values > 10 or < .1 as strong evidence for $H_1$ and $H_0$ respectively.

```python
# Test bounds in each session
def get_windowed_mi(x,y,nw):
    mis = np.zeros(nw+1)
    for w in np.arange(1,nw+1):
        xw,yw = ut.get_windowed_x(x,y,w=w)
        mis[w] = ut.mutual_inf_nsb(xw,yw,[2**w,2])
    return(mis)

# Get all trials in all three sessions that match low and high hazard rates
nw = 10

adat_mis_h1 = get_windowed_mi(adat_trials['X'].loc[adat_trials['H'] == .01]-1,
                              adat_trials['Z'].loc[adat_trials['H'] == .01],nw)

adat_mis_h9 = get_windowed_mi(adat_trials['X'].loc[adat_trials['H'] == .99]-1,
                              adat_trials['Z'].loc[adat_trials['H'] == .99],nw)
plt.plot(np.arange(nw+1),adat_mis_h1,'-o')
plt.plot(np.arange(nw+1),adat_mis_h9,'-o')
```

```python
%%time
def get_windowed_bound(x,y,w,mb=50,nb=2000,p=8):
    xw,yw = ut.get_windowed_x(x,y,w=w)
    ipw,ifw,betasw = EmpiricalBottleneck(xw,yw,minsize=True,processes=p,maxbeta=mb,numbeta=nb).get_empirical_bottleneck() 
    return((ipw,ifw))

ipw_h1,ifw_h1 = get_windowed_bound(adat_trials['X'].loc[adat_trials['H'] == .01]-1,
                                   adat_trials['Z'].loc[adat_trials['H'] == .01],8)
ipw_h1_1back,ifw_h1_1back = get_windowed_bound(adat_trials['X'].loc[adat_trials['H'] == .01]-1,
                                               adat_trials['Z'].loc[adat_trials['H'] == .01],1)

ipw_h9,ifw_h9 = get_windowed_bound(adat_trials['X'].loc[adat_trials['H'] == .99]-1,
                                   adat_trials['Z'].loc[adat_trials['H'] == .99],8)
ipw_h9_1back,ifw_h9_1back = get_windowed_bound(adat_trials['X'].loc[adat_trials['H'] == .99]-1,
                                               adat_trials['Z'].loc[adat_trials['H'] == .99],1)
```

```python
def get_windowed_mi_subs(sdat_all,nw):
    subs = pd.unique(sdat_all['Subject'])
    ip_sub = {}
    if_sub = {}
    for subi,sub in enumerate(subs):
        sdat = sdat_all[sdat_all['Subject'] == sub]
        ip_sub[sub] = get_windowed_mi(sdat['X']-1,sdat['Resp'],nw)
        if_sub[sub] = ut.mutual_inf_nsb(sdat['Resp'],sdat['Z'],[2,2])
    return(ip_sub,if_sub)

ips_sub_h1,if_sub_h1 = get_windowed_mi_subs(adat_h1,10)
ips_sub_h9,if_sub_h9 = get_windowed_mi_subs(adat_h9,10)
```

```python
%%time
def get_bootstrapped_samples(dat,wips,nboot):
    subs = pd.unique(dat['Subject'])
    ip_boot_mi = {}
    if_boot_mi = {}
    for subi,sub in enumerate(subs):
        sdat = dat[dat['Subject'] == sub]
        w = wips[sub].argmax()
        x = np.array(sdat['X']-1)
        r = np.array(sdat['Resp'])
        y = np.array(sdat['Z'])
        xw,rw = ut.get_windowed_x(x,r,w=(w+1))
        ip_boot_mi[sub] = np.zeros(nboot)
        if_boot_mi[sub] = np.zeros(nboot)
        for boot in np.arange(nboot):
            idx = np.random.choice(np.arange(len(xw)),size=len(xw),replace=True)
            ip_boot_mi[sub][boot] = ut.mutual_inf_nsb(xw[idx],rw[idx],[2**(w+1),2])
            if_boot_mi[sub][boot] = ut.mutual_inf_nsb(r[idx],y[idx],[2,2])
    return(ip_boot_mi,if_boot_mi)

nboot = 1000
ip_h1_boot, if_h1_boot = get_bootstrapped_samples(adat_h1,ips_sub_h1,nboot)
ip_h9_boot, if_h9_boot = get_bootstrapped_samples(adat_h9,ips_sub_h9,nboot)
```

```python
def deltaBound(ib_ipast,ib_ifuture,p_ipast,p_ifuture):
    ''' Function to calculate distance from the bound between an empirical IB and participant predictive info
    ib_ipast: ipast of empirical IB (x of convex hull)
    ib_ifuture: ifuture of empirical IB (y of convex hull)
    p_ipast: participant ipast (uncorrected)
    p_ifuture: participant ifuture (uncorrected)
    
    Returns participant ifuture minus the empirical bound (more negative = farther away from the bound)
    '''
    ind = np.argwhere(np.array(ib_ipast) > p_ipast)[0][0]
    slp = (ib_ifuture[ind]-ib_ifuture[ind-1])/(ib_ipast[ind]-ib_ipast[ind-1])
    intercept = ib_ifuture[ind]-(slp*ib_ipast[ind])
    #Return distance between participant Ifuture and interpolated bound - higher 
    return p_ifuture - ((p_ipast*slp)+intercept)

dbs1 = np.zeros(len(ip_mean_h1))
dbs9 = np.zeros(len(ip_mean_h9))
for i in np.arange(len(ip_mean_h1)):
    dbs1[i] = deltaBound(ipw_h1,ifw_h1,ip_mean_h1[i],if_mean_h1[i])
    dbs9[i] = deltaBound(ipw_h9,ifw_h9,ip_mean_h9[i],if_mean_h9[i])
plt.figure(1,(15,10))
plt.subplot(2,2,1)
plt.plot(ip_mean_h1,dbs1,'o',color=cmap(0))
from scipy.stats import spearmanr
print(spearmanr(ip_mean_h1,dbs1))

plt.subplot(2,2,2)
plt.plot(ip_mean_h9,dbs9,'o',color=cmap(1))
print(spearmanr(ip_mean_h9,dbs9))

plt.subplot(2,2,3)
ip1b = ipw_h1_1back[ipw_h1_1back<max(ipw_h1)]
if1b = ifw_h1_1back[ifw_h1_1back<max(ifw_h1)]
h1_1back_dfb = np.array([deltaBound(ipw_h1,ifw_h1,ip1b[i],if1b[i]) for i in np.arange(len(ip1b))])
plt.plot(ip1b,h1_1back_dfb,'o',color=cmap(0))

plt.subplot(2,2,4)
ip9b = ipw_h9_1back[ipw_h9_1back<max(ipw_h9)]
if9b = ifw_h9_1back[ifw_h9_1back<max(ifw_h9)]
h9_1back_dfb = np.array([deltaBound(ipw_h9,ifw_h9,ip9b[i],if9b[i]) for i in np.arange(len(ip9b))])
plt.plot(ip9b,h9_1back_dfb,'o',color=cmap(1))
```
